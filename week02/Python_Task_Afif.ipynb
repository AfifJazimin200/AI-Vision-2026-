```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sediakan data awal (array markah dan nama pelajar contoh)
markah = np.array([80, 75, 90, 85, 70])
nama_pelajar = ['Amir', 'Balqis', 'Chong', 'Devi', 'Ethan']

# ==========================================
# A. Paparkan Statistik Menggunakan NumPy
# ==========================================
print("=== A. RINGKASAN STATISTIK MARKAH ===")
print(f"Semua Markah    : {markah}")
print(f"Markah Tertinggi: {np.max(markah)}")
print(f"Markah Terendah : {np.min(markah)}")
print(f"Purata Markah   : {np.mean(markah)}")
print(f"Jumlah (Sum)    : {np.sum(markah)}")
print("-" * 40)

# ==========================================
# B. Bina DataFrame Menggunakan Pandas
# ==========================================
df = pd.DataFrame({
    'Nama Pelajar': nama_pelajar,
    'Markah': markah
})

print("\n=== B. DATAFRAME PRESTASI PELAJAR ===")
print(df.to_string(index=False))
print("-" * 40)

# ==========================================
# C. Penghasilan Bar Chart Menggunakan Matplotlib
# ==========================================
plt.figure(figsize=(8, 5))

# Membina bar chart dengan warna bertema akademik
colors = ['#2b5c8f', '#4682b4', '#5f9ea0', '#20b2aa', '#3cb371']
plt.bar(df['Nama Pelajar'], df['Markah'], color=colors, edgecolor='#1e293b', width=0.6)

# Menetapkan tajuk dan label paksi
plt.title('Prestasi Markah Pelajar KV Sepang', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Nama Pelajar', fontsize=12, fontweight='bold', labelpad=10)
plt.ylabel('Markah', fontsize=12, fontweight='bold', labelpad=10)

# Kemasan tambahan pada graf
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Memaparkan graf secara visual
plt.show()
```
